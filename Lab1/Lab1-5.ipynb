2. Какой медианный и средний возраста пассажиров

````python
# Здесь и далее data это df без отсутствующих возрастов
data = df[df['Age'].notna()]
median_age = data['Age'].median()
mean_age = data['Age'].mean()
print(median_age)
print(mean_age)
````

3. Посчитайте процент выживаемости детей(до 16 лет) и взрослых

````python
young = data[data.Age <= 16]
print(len(young[young['Survived'] == 1]) / len(young) * 100)
````

4. Верна ли гипотеза, что женщины и дети сажались в шлюпки первыми и выживали больше?

````python
d = data[(data.Age <= 16) | (data.Sex == 'female')]
d2 = data[(data.Age > 16) | (data.Sex == 'male')]
d_percentage = len(d[d.Survived == 1]) / len(d) * 100 
d2_percentage = len(d2[d2.Survived == 1]) / len(d2) * 100 
print(d_percentage > d2_percentage)
````

5. Зависит ли выживаемость от класса обслуживания?

````python
classes = df.Pclass.unique()
classes_survivability = np.zeros(len(classes))
for c in classes:
    classes_survivability[c - 1] = len(df[(df.Pclass == c) & (df.Survived == 1)]) / len(df[df.Pclass == 1]) * 100
classes_survivability
# Между классами обслуживания существенная разница в проценте выживания, значит существует зависимость
````

6. Посчитайте средний возраст умерших женщин и мужчин

7. Различается ли процент выживаемости пассажиров с хотя бы одним родственником на борту и одиночек?

````python
has_relative = df[(df.SibSp > 0) | (df.Parch > 0)]
has_no_relative = df[(df.SibSp == 0) | (df.Parch == 0)]
print(len(has_relative[has_relative.Survived == 1]) / len(has_relative) * 100)
print(len(has_no_relative[has_no_relative.Survived == 1]) / len(has_no_relative) * 100)
# Да, различаются. Процет выжиываемости одиночек ниже
````

8. Различается ли средняя стоимость билета у умерших и выживших пассажиров?

````python
d = df[df.Fare.notna()]
surivived_mean = d[d.Survived == 1].Fare.mean()
dead_mean = d[d.Survived == 0].Fare.mean()
print(surivived_mean)
print(dead_mean)
# Да, различаются. У мертвых цена ниже примерно в два раза
````

10. Подтвердите или опровергните гипотезу: молодым мужчинам(от 18 лет) выжить легче, чем более взрослым

````python
young_men = data[(data.Age >= 18) & (data.Age < 30) & (data.Sex == 'male')]
old_men = data[(data.Age >= 30) & (data.Sex == 'male')]
young_percentage = len(young_men[young_men.Survived == 1]) / len(young_men) * 100
old_percentage = len(old_men[old_men.Survived == 1]) / len(old_men) * 100
print(young_percentage > old_percentage)
# Гипотеза ложна
````

5. Визуализируйте распределение долей выживших среди мужчин, женщин и детей(до 16 лет)

````python
male = data[(data.Sex == 'male') & (data.Age > 16)]
female = data[(data.Sex == 'female') & (data.Age > 16)]
children = data[data.Age <= 16]
groups = (male, female, children)
group_names = ('Мужчины', 'Женщины', 'Дети')

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for i in range(len(groups)):
    axs[i].pie((groups[i].Survived.mean(), 1 - groups[i].Survived.mean()),
               labels=('Выжившие', 'Не выжившие'),
               autopct='%1.1f%%')
    axs[i].set_title(group_names[i])

plt.suptitle('Доля выживших среди мужчин, женщин и детей')
plt.show()
````

\* Задание на доп. балл

````python
survival_count = df.groupby(['Pclass'])['Survived'].sum()
survival_count_by_class = df.groupby(['Pclass', 'Sex'])['Survived'].sum()

plt.figure(figsize=(8,8))
size = 0.3
cmap = plt.colormaps["tab20c"]
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap([1, 2, 5, 6, 9, 10])

plt.pie(survival_count,
        radius=1,
        colors=outer_colors,
        autopct='%1.1f%%',
        pctdistance=0.85,
        labels=[f'{i} класс' for i in range(1,4)],
        wedgeprops=dict(width=size, edgecolor='w'))

plt.pie(survival_count_by_class,
        radius=1-size,
        colors=inner_colors,
        autopct='%1.1f%%',
        pctdistance=0.78,
        labels=np.array(['Ж', 'М']*3).flatten(),
        labeldistance=0.48,
        wedgeprops=dict(width=size, edgecolor='w'))

plt.title('Выживаемость мужчин и женщин в разных классах')
plt.show()
````
